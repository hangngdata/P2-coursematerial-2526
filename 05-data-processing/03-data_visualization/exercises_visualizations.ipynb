## 1. Counting birds

### <u> Counting birds part 1 </u>

Every year there is a day where people are encouraged to count the birds in their garden. Joseph tries to participate every year. Below you see the data of Joseph's observations. Let's make a nice plot of the data, also include the total amount of birds spotted.

```python
# Sample data
data = {
    'year': ['2020', '2021', '2022', '2023', '2024'],
    'Pigeons': [33, 52, 26, 40, 38],
    'Sparrows': [54, 48, 66, 59, 61],
    'others': [104, 98, 88, 92, 101]
}
```

### Inspecting the data

Load and inspect the data.

```python
# Load and inspect
df = pd.DataFrame(data)
df.head()
```

```python
df.dtypes
```

### Plotting the pigeon count

Plot the number of counted pigeons over the year, following these specifications:

* Line graph
* Size of the figure: (10,6)
* Full blue line, data point marked with a circle, linewidth 2, marker size 8
* Provide the necessary labels in order to interpret the graph and show a legend
* Provide a meaningfull title
* Show a grid

```python
#Plotting the Pigeon count
plt.figure(figsize=(10, 6))

plt.plot(df['year'], df['Pigeons'], marker='o', color='blue', linestyle='-', linewidth=2, markersize=8, label='Pigeons')

plt.title("Pigeon Count over Years")

plt.xlabel('Year')
plt.ylabel('Count')

plt.legend()
plt.grid(True)

plt.show()
```

### Plotting the sparrow count

Plot the number of sparrows Joseph counted over the years, adhere to the specifications listed below:

* Line graph
* linestyle: dotted
* color of the graph: Green
* marker: a triangle pointing to the right
* the edge of the markers should be red with width 1
* Corresponding labels and titles

Ensure the plot of the Pigeon count is incorporated as well.


```python
# Plotting the sparrow count
plt.figure(figsize=(10, 6))

plt.plot(df['year'], df['Pigeons'], marker='o', color='blue', linestyle='-', linewidth=2, markersize=8, label='Pigeons')
plt.plot(df['year'], df['Sparrows'], marker='>', color='green', mec='r', linestyle='dotted', linewidth=1, markersize=8, label='Sparrows')

plt.title("Pigeon and Sparrow Count over Years")

plt.xlabel('Year')
plt.ylabel('Count')

plt.legend()
plt.grid(True)

plt.show()
```

### Plotting the other birds

Add a graph line for all other birds counted. You can use a styling of your choice.

```python
# Plotting the other birds count
plt.figure(figsize=(10, 6))

plt.plot(df['year'], df['others'], marker='o', color='blue', linestyle='-', linewidth=2, markersize=8, label='Other birds')

plt.title("Other Birds Count over Years")

plt.xlabel('Year')
plt.ylabel('Count')

plt.legend()
plt.grid(True)

plt.show()
```

### <u> Counting birds part 2 </u>

### Plotting the total of birds counted

Add a column to the dataframe containing the total amount of birds counted that year.


```python
# Total of birds
df['Total'] = df[['Pigeons', 'Sparrows', 'others']].sum(axis=1)
df
```



Add a **barplot** for the total of birds counted that day on top of the already created plots of the bird species. Some specifications:
* Bar width = 0.4
* Bar color = purple
* Use corresponding labels 

Use corresponding labels 

```python
# Write your code here
plt.figure(figsize=(10, 6))
bar_width = 0.4

index = range(len(df['year']))

plt.plot(df['year'], df['Pigeons'], marker='o', color='blue', linestyle='-', linewidth=2, markersize=8, label='Pigeons')
plt.plot(df['year'], df['Sparrows'], marker='>', color='green', linestyle='-', linewidth=2, markersize=8, label='Sparrows')
plt.plot(df['year'], df['others'], marker='*', color='yellow', linestyle='-', linewidth=2, markersize=8, label='Others')
plt.bar(index, df['Total'], bar_width, color='purple', label='Total')

plt.xticks(index, df['year'])

plt.legend()

plt.grid(axis='y')
plt.show()
```

### <u> Counting birds part 3 </u>

### Bar chart of total
Display the data  using a bar chart (total) and a stacked bar chart (pigeons, sparrows, others). Ensure the total is displayed on the right side and the stacked bar on the left.
Make it informative (and pretty).

```python
# Your code
plt.figure(figsize=(10, 6))
bar_width = 0.4

index = range(len(df['year']))

plt.bar(index, df['Pigeons'], bar_width, color='red', label='Pigeons')
plt.bar(index, df['Sparrows'], bar_width, bottom=df['Pigeons'], color='green', label='Sparrows')
plt.bar(index, df['others'], bar_width, bottom=df['Pigeons']+df['Sparrows'], color='blue', label='Others')
plt.bar([i + bar_width for i in index], df['Total'], bar_width, color='purple', label='Total')

plt.xticks([i + bar_width/2 for i in index], df['year'])

plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Bird Count over Year')

plt.legend()

plt.grid(False)
plt.show()
```

### <u> Counting birds part 4 </u>

### Grouped bar chart

Display the birds data using a grouped bar chart. Place the count of the pigeons and sparrows on the left of the x-tick, others and total on the right. Ensure the bars do not overlap.

```python
# Your code
plt.figure(figsize=(10, 6))
bar_width = 0.2

index = range(len(df['year']))

plt.bar([i - bar_width*1.5 for i in index], df['Pigeons'], bar_width, color='red', label='Pigeons')
plt.bar([i - bar_width*0.5 for i in index], df['Sparrows'], bar_width, color='green', label='Sparrows')
plt.bar([i + bar_width*0.5 for i in index], df['others'], bar_width, color='blue', label='Others')
plt.bar([i + bar_width*1.5 for i in index], df['Total'], bar_width, color='purple', label='Total')

plt.xticks(index, df['year'])

plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Bird Count over Year')

plt.legend()
plt.show()
```
